### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Creamos un figure con 4 axes
fig,ax = plt.subplots(2,2, figsize = (16,16));
```

```python
# Introducimos los 4 diagramas

# DIAGRAMA DE BARRAS

# Ordenamos los valores para tener los más poblados al principio
df.sort_values(by = "population_total",inplace = True, ascending = False)

# Cogemos los valores de las primeras ciudades y tambien los que queremos que muestre
ciudades = ["LosAngeles", "SanDiego", "Sacramento","Anaheim","SantaAna"]
ciudades2 =  ["Los Angeles", "San Diego", "Sacramento","Anaheim","Santa Ana"]
condicion = df["city"].isin(ciudades)

# Me quedo con los valores de poblacion
poblacion = df.loc[condicion,"population_total"]

# Marcamos el tamaño de texto
plt.rcParams['font.size'] = 8

#Genero el diagrama de barras
ax[0,0].bar(ciudades2, poblacion/1000, color = ["blue","grey","salmon","olive","brown"], alpha = 0.5)
ax[0,0].set_title("Ciudades más pobladas de California",size = 10)
ax[0,0].set_xlabel("Ciudades", size = 8)
ax[0,0].set_ylabel("Habitantes (miles)", size = 8)
ax[0,0].set_yticks([1000,2000,3000,4000,5000])

# Metemos los textos que necesito
for i in range(5):
    condicion = df["city"] == ciudades[i]
    valor = df.loc[condicion,"population_total"].to_list()[0]
    altura = df.loc[condicion,"population_total"]
    altura = float (altura.iloc[0])
    ax[0,0].text(ciudades2[i],float(df.loc[condicion,"population_total"].iloc[0])/1000+150,valor/1000, ha = "center",size = 8)



# HISTOGRAMA

# Marcamos el tamaño de texto
plt.rcParams["font.family"] = "DejaVu Sans"
plt.rcParams['font.size'] = 8

#Creamos el histograma
ax[0,1].hist(df["area_total_km2"],bins =50, label = "hist Area",color = "steelblue")
ax[0,1].legend()
ax[0,1].set_xticks(list(range(0,1400,200)))
ax[0,1].set_yticks([0,50,100,150,200])

# usamos un diccionario para cambiar el estilo de ciertos puntos
font = {'family': 'Bradley Hand itc',
        'color':'black',
        'size': 18
}

# Cambiamos el tipo de letra de los ejes
ax[0,1].set_xlabel("Superficie ($Km^2$)", fontdict=font )
ax[0,1].set_ylabel("Numero de ciudades", fontdict=font)




# DIAGRAMA DE DISPERSION

# Establezco un diccionario para los casos que necesitan establecer el tipo de texto
font = {'family': 'DejaVu Sans',
        'color':'black',
        'size': 8
}
# Marcamos el tamaño de texto
plt.rcParams["font.family"] = "DejaVu Sans"
plt.rcParams['font.size'] = 8

# Genero el diagrama de dispersion
ax[1,0].scatter(df["area_water_km2"],df["population_total"]/1000, color = "green", alpha = 0.5)
ax[1,0].set_title("Poblacion por Km2 de agua",size = 10)
ax[1,0].set_xlabel("Area de agua en la ciudad ($Km^2$)", fontdict=font )
ax[1,0].set_ylabel("Población (miles de habitantes)", fontdict=font)

# Establezco el tipo de flecha
props_flecha = {
    "arrowstyle": "->"
}
# Genero los valores de la posicion de la flecha
maximo_pop_y = df["population_total"].max()/1000
maximo_pop_x = df.loc[df["population_total"]==df["population_total"].max(),"area_water_km2"].to_list()[0]
maximo_wat_x = df["area_water_km2"].max()
maximo_wat_y = df.loc[df["area_water_km2"]==maximo_wat_x,"population_total"].to_list()[0]/1000

# Incluimos la anotación
ax[1,0].annotate(text="outliers", xy = (maximo_pop_x,maximo_pop_y),xytext = (60,3600),arrowprops = props_flecha)
ax[1,0].annotate(text="        ", xy = (maximo_wat_x,maximo_wat_y),xytext = (60,3600),arrowprops = props_flecha)





# BOXPLOT

# Establezco un diccionario para los casos que necesitan establecer el tipo de texto
font = {'family': 'DejaVu Sans',
        'color':'black',
        'size': 8
}
# Marcamos el tamaño de texto
plt.rcParams["font.family"] = "DejaVu Sans"
plt.rcParams['font.size'] = 8

# Dibujamos el objeto caja y nos lo quedamos en una variable para referenciarlo
caja = ax[1,1].boxplot(df["elevation_m"],
                   whis = 1.5, 
                   vert = False, 
                   patch_artist = True,
                   ); 

#Cambiamos el color a la caja
print(caja["medians"])
for patch in caja["boxes"]:
    patch.set_facecolor("red")

for line in caja["medians"]:
    line.set_color("black")

# Ponemos el título
ax[1,1].set_title("Diagrama de caja de la Altitud de las ciudades de California",size = 10)


fig.savefig("./img/Practica_obligatoria_Sprint8_Ud1_figura.jpg")
fig



```
